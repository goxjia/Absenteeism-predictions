```python
import pandas as pd
```

```python
raw_data = pd.read_csv('Absenteeism_data.csv')
df = raw_data.copy()
```

```python
df.head()
```

```python
df.columns.values
```

## 删除ID列

```python
df = df.drop(['ID'], axis=1)
df.head()
```

## Date列的整理

```python
type(df['Date'][0])
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df.head()
```

```python
type(df['Date'][0])
```

```python
# 增加月份以及dow列
df['Month'] = df['Date'].dt.month # 一月是1
df['Day of week'] = df['Date'].dt.dayofweek # 周一是0
```

```python
# 删除无用的Date列
df = df.drop(['Date'], axis=1)
```

```python
# 检查点
df_date_mod = df.copy()
df_date_mod.head()
```

## 缺勤原因dummies化以及分类整理

```python
# 最多显示20行
pd.options.display.max_rows = 20
```

```python
reason_columns = pd.get_dummies(df_date_mod['Reason for Absence'])
reason_columns
```

```python
type(reason_columns)
```

```python
reason_columns['check'] = reason_columns.sum(axis=1)
```

```python
reason_columns.head()
```

```python
reason_columns['check'].unique()
```

```python
# 将原因分为四类
Reason_1 = reason_columns.loc[:, :14].max(axis=1).astype(int)
Reason_2 = reason_columns.loc[:, 15:17].max(axis=1).astype(int)
Reason_3 = reason_columns.loc[:, 18:21].max(axis=1).astype(int)
Reason_4 = reason_columns.loc[:, 22:28].max(axis=1).astype(int)
```

```python
# 将四类原因与原始dataframe合并，并删除本来的原因列
df_date_mod = pd.concat([df_date_mod, Reason_1, Reason_2, Reason_3, Reason_4], axis=1)
df_date_mod = df_date_mod.drop(['Reason for Absence'], axis=1)
```

```python
# 存档点
df_reason_mod = df_date_mod.copy()
df_reason_mod.head()
```

## 处理Education列

```python
df_reason_mod['Education'].value_counts()
```

```python
# 将Education为1（高中学历）的转化为0，其他为1
df_reason_mod['Education'] = df_reason_mod['Education'].map({1:0, 2:1, 3:1, 4:1})
df_reason_mod.head(20)
```

## 更改列名及排序

```python
# 更改列名
df_reason_mod = df_reason_mod.rename(columns={0:'Reason_1', 1:'Reason_2', 2:'Reason_3', 3:'Reason_4'})
df_reason_mod.head()
```

```python
# 更改列的顺序
columns_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month',
       'Day of week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df_reordered = df_reason_mod[columns_reordered]
df_reordered.head()
```

## 导出

```python
df_reordered.to_csv( 'Absenteeism_preprocessed.csv', index=False)
```
