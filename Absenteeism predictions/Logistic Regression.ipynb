## 环境准备及导入数据

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Absenteeism_preprocessed.csv')
df.head()
```

## 构建目标变量

```python
# 如果缺勤时间大于该列的中位数，即为过度缺勤
target = np.where(df['Absenteeism Time in Hours'] > df['Absenteeism Time in Hours'].median(), 1, 0)
target
```

```python
# 使用中位数设置目标值可以平衡数据集
target.sum() / target.shape[0]
```

```python
df['Excessive Absenteeism'] = target
df.head()
```

```python
# check point
data_with_target = df.drop(columns=['Day of the Week', 'Distance to Work', 'Daily Work Load Average', 'Absenteeism Time in Hours'], axis=1)
data_with_target.head()
```

## 特征工程

```python
data_with_target.columns.values
```

```python
unscaled_features = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value',
       'Transportation Expense', 'Age', 'Body Mass Index', 'Education',
       'Children', 'Pet']
unscaled_inputs = data_with_target[unscaled_features]
```

```python
unscaled_inputs
```

## 数据标准化

```python
from sklearn.preprocessing import StandardScaler
```

```python
absenteeism_scaler = StandardScaler()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler
class CustomScaler(BaseEstimator, TransformerMixin): 
    
    def __init__(self, columns, copy=True, with_mean=True, with_std=True):
        self.scaler = StandardScaler(copy=copy, with_mean=with_mean, with_std=with_std)
        self.columns = columns
        # Store the parameters as instance attributes
        self.copy = copy
        self.with_mean = with_mean
        self.with_std = with_std
    
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```

```python
unscaled_inputs.columns.values
```

```python
columns_to_omit = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Education']
```

```python
columns_to_scale = [column for column in unscaled_inputs.columns.values if column not in columns_to_omit]
columns_to_scale
```

```python
absenteeism_scaler = CustomScaler(columns_to_scale)
```

```python
# fit the data (calculate mean and standard deviation); they are automatically stored inside the object 
absenteeism_scaler.fit(unscaled_inputs)
```

```python
# the scaled_inputs are now an ndarray, because sklearn works with ndarrays
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
scaled_inputs
```

```python
scaled_inputs.shape
```

## 数据集划分

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, target, random_state=20, train_size=0.8)
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

## 模型训练

```python
# import the LogReg model
from sklearn.linear_model import LogisticRegression
# import the 'metrics' module, which includes important metrics we may want to use
from sklearn import metrics
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_train)
predictions
```

```python
predictions == y_train
```

```python
# 手动计算精确度
accuracy = np.sum((predictions == y_train)) / y_train.shape[0]
accuracy
```

```python
# 调用函数计算精准度
model.score(x_train, y_train)
```

## 模型解释

```python
# 计算截距
intercept = model.intercept_[0]
intercept
```

```python
# 计算系数
model.coef_
```

```python
features = list(scaled_inputs.columns.values)
features
```

```python
summary_table = pd.DataFrame(columns=['features'], data=features)
```

```python
summary_table['coefficients'] = np.transpose(model.coef_)
```

```python
# 特征重要性表格
summary_table
```

```python
# 将截距项加入表格
summary_table.index += 1
summary_table.loc[0] = ['Intercept', intercept]
```

```python
summary_table = summary_table.sort_index()
summary_table
```

```python
summary_table['odds_ratio'] = np.exp(summary_table['coefficients'])
```

```python
summary_table
```

```python
# 按优势比降序排序
summary_table = summary_table.sort_values(by='odds_ratio', ascending=False)
summary_table
```

## 测试模型

```python
# 计算精准度
model.score(x_test, y_test)
```
